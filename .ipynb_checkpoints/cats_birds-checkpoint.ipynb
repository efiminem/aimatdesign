## Linear

```python
class Linear(nn.Module):
    def __init__(self):
        super(Linear, self).__init__()
        self.flatten = Flatten()
        self.linear = nn.Linear(75 * 100 * 3, 2)
    def forward(self, images):
        flattened_images = self.flatten(images)
        answer = self.linear(flattened_images)
        return answer
```

## Simple dense net

```python
class SimpleDense(nn.Module):
    def __init__(self):
        super(SimpleDense, self).__init__()
        self.flatten = Flatten()
        self.first_linear = nn.Linear(75 * 100 * 3, 100)
        self.second_linear = nn.Linear(100, 2)
        self.relu = nn.ReLU()
    def forward(self, X):
        X = self.flatten(X)
        X = self.first_linear(X)
        X = self.relu(X)
        X = self.second_linear(X)
        return X
```

```python
model = SimpleDense()
model.cuda()
best_model = perform_training(model, images_train, labels_train, images_val, labels_val, n_epochs = 50, lr=1e-5)
```

## Simple convolution net

```python
class SimpleConv(nn.Module):
    def __init__(self):
        super(SimpleConv, self).__init__()
        self.flatten = Flatten()
        self.conv = nn.Conv2d(3, 5, (3, 3))
        self.pool = nn.MaxPool2d((2, 2))
        self.relu = nn.ReLU()
        
        self.linear = nn.Linear(8820, 2)
       
    def forward(self, X):
        X = self.conv(X)
        X = self.pool(X)
        X = self.relu(X)
        X = self.flatten(X)
        X = self.linear(X)
        return X
```

```python
model = SimpleConv()
model.cuda()
best_model = perform_training(model, images_train, labels_train, images_val, labels_val, n_epochs = 50, lr=1e-4)
```
