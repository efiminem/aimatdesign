### Simple dense net

```python
class SimpleDense(nn.Module):
    def __init__(self):
        super(SimpleDense, self).__init__()
        self.flatten = Flatten()
        self.first_linear = nn.Linear(75 * 100 * 3, 100)
        self.second_linear = nn.Linear(100, 2)
        self.relu = nn.ReLU()
    def forward(self, X):
        X = self.flatten(X)
        X = self.first_linear(X)
        X = self.relu(X)
        X = self.second_linear(X)
        return X
```

```python
class SimpleConv(nn.Module):
    def __init__(self):
        super(SimpleConv, self).__init__()
        self.flatten = Flatten()
        self.conv = nn.Conv2d(3, 5, (3, 3))
        self.pool = nn.MaxPool2d((2, 2))
        self.relu = nn.ReLU()
        
        self.linear = nn.Linear(8820, 2)
       
    def forward(self, X):
        X = self.conv(X)
        X = self.pool(X)
        X = self.relu(X)
        X = self.flatten(X)
        X = self.linear(X)
        return X
```

## Part II. Materials properties

Here using similar principles we will construct the neural network for predicting two properties: 
<b><i>Young's modulus</i></b> and <b><i>Poisson's ratio</i></b>. These properties fully characterize the elastic behavior of the material.

As inputs to our network we have crystal structures. All structures are respresented by CIF format, which contains information about atoms and their coordinates.

```python
dataset = CIFData(root_dir = "elastic",
                  max_num_nbr = 12,
                  radius=8,
                  dmin=0,
                  step=0.2)
```

```python
orig_atom_fea_len = dataset[0][0][0].shape[-1]
nbr_fea_len = dataset[0][0][1].shape[-1]
print("Number of structures: {}".format(len(dataset)))
print("Number of features describing one atom: {}".format(orig_atom_fea_len))
print("Number of features describing neighbours: {}".format(nbr_fea_len))
```

```python
normalizer = Normalizer(collate_pool([dataset[i] for i in range(500)])[1])
```

```python
normalizer.mean
```

```python
train_val_ratio = 0.6
indices = list(range(len(dataset)))
train_sampler = SubsetRandomSampler(indices[:int(train_val_ratio*len(dataset))])
val_sampler = SubsetRandomSampler(indices[int(train_val_ratio*len(dataset)):])
train_loader = DataLoader(dataset, batch_size=16, sampler=train_sampler, collate_fn=collate_pool)
val_loader = DataLoader(dataset, batch_size=16, sampler=val_sampler, collate_fn=collate_pool)
print("Number of structures in train set: {}".format(len(train_sampler)))
print("Number of structures in validation set: {}".format(len(val_sampler)))
```

```python
class CrystalGraphConvNet(nn.Module):
    """
    Create a crystal graph convolutional neural network for predicting total
    material properties.
    """
    def __init__(self, orig_atom_fea_len, nbr_fea_len,
                 atom_fea_len=60, h_fea_len=50):
        """
        Initialize CrystalGraphConvNet.
        Parameters
        ----------
        orig_atom_fea_len: int
          Number of atom features in the input.
        nbr_fea_len: int
          Number of bond features.
        atom_fea_len: int
          Number of hidden atom features in the convolutional layers
        h_fea_len: int
          Number of hidden features after pooling
        """
        super(CrystalGraphConvNet, self).__init__()
        self.embedding = nn.Linear(orig_atom_fea_len, atom_fea_len)
        self.conv = GraphConvLayer(atom_fea_len=atom_fea_len, nbr_fea_len=nbr_fea_len)
        self.pooling = GraphPooling()
        self.conv_to_fc = nn.Linear(atom_fea_len, h_fea_len)
        self.fc_out = nn.Linear(h_fea_len, 2)

    def forward(self, atom_fea, nbr_fea, nbr_fea_idx, crystal_atom_idx):
        """
        Forward pass
        N: Total number of atoms in the batch
        M: Max number of neighbors
        N0: Total number of crystals in the batch
        Parameters
        ----------
        atom_fea: Variable(torch.Tensor) shape (N, orig_atom_fea_len)
          Atom features from atom type
        nbr_fea: Variable(torch.Tensor) shape (N, M, nbr_fea_len)
          Bond features of each atom's M neighbors
        nbr_fea_idx: torch.LongTensor shape (N, M)
          Indices of M neighbors of each atom
        crystal_atom_idx: list of torch.LongTensor of length N0
          Mapping from the crystal idx to atom idx
        Returns
        -------
        prediction: nn.Variable shape (N, )
          Atom hidden features after convolution
        """
        atom_fea = self.embedding(atom_fea)
        atom_fea = self.conv(atom_fea, nbr_fea, nbr_fea_idx)
        crys_fea = self.pooling(atom_fea, crystal_atom_idx)
        crys_fea = self.conv_to_fc(crys_fea)
        out = self.fc_out(crys_fea)
        
        return out
```

```python
model = CrystalGraphConvNet(orig_atom_fea_len, nbr_fea_len)
model.cuda()
```

```python
best_model = perform_graph_training(model, train_loader, val_loader, normalizer, n_epochs = 100, lr=0.1)
```

## Part III. Materials search

We call the material hard if is has high values of both <i><b>hardness</b></i> and <i><b>fracture toughness</b></i>.

Roughly speaking, <i><b>hardness</b></i> characterises the material resistence to local deformation while <i><b>fracture toughness</b></i> measures the ability of material to resist fracture (imagine that first is produced by wide indenter and the second is produced by needle). There is no exect formula for both of them, however, they can be calculated approximately.

Hardness:
$$H = \frac{(1 - 8.5\nu + 19.5\nu^2)E}{1 - 7.5\nu + 12.2\nu^2 + 19.6\nu^3}$$

```python
def H(E, nu):
    return 0.096*E*(1 - 8.5*nu + 19.5*nu**2)/(1 - 7.5*nu + 12.2*nu**2 + 19.6*nu**3)
```

Fracture toughness:
$$K_{\text{IC}} = \alpha_0^{1/2} V_0^{1/6} \left( \frac{(1 - 13.7\nu + 48.6\nu^2)E}{1 - 15.2\nu + 70.2\nu^2 - 81.5\nu^3} \right)^{3/2},$$<br>
where $\alpha_0 \approx 8840\text{ GPa}$ depends on the chemical bonding in the material and has units of pressure, $V_0$ is the volume per structure

```python
def K(V0, E, nu):
    return (10**-2)*(8840**-0.5)*(V0**(1/6))*\
        (E*(1 - 13.7*nu + 48.6*nu**2)/(1 - 15.2*nu + 70.2*nu**2 - 81.5*nu**3))**1.5
```
